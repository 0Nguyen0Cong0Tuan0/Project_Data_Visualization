```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv(filepath_or_buffer='vn_weather_2019_2024.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(columns=['UV Index Max', 'UV Index Clear Sky Max', 'Precipitation Probability Max (%)', 'Showers (mm)', 'Snowfall (mm)'], inplace=True)
```

```python
df['Date'] = pd.to_datetime(arg=df['Date'], format='%Y-%m-%d')
df['Sunrise'] = pd.to_datetime(arg=df['Sunrise'], format='%Y-%m-%dT%H:%M')
df['Sunset'] = pd.to_datetime(arg=df['Sunset'], format='%Y-%m-%dT%H:%M')
```

```python
df.info()
```

```python
df['Province'].unique()
```

```python
df = df[df['Province'] != 'Ha Tay']
df['Province'] = np.where(
    df['Province'] == 'Dac Lak', 'Dak Lak', np.where(
        df['Province'] == 'Ho Chi Minh', 'Thanh pho Ho Chi Minh', df['Province']
    )
)
```

```python
df['Province'].nunique()
```

```python
df['Province'].value_counts().head()
```

```python
dien_bien_data = df[df['Province'] == 'Lai Chau'].iloc[:, 1:]
dien_bien_data['Province'] = 'Dien Bien'
dien_bien_data.head()
```

```python
dak_nong_data = df[df['Province'] == 'Dak Lak'].iloc[:, 1:]
dak_nong_data['Province'] = 'Dak Nong'
dak_nong_data.head()
```

```python
# Provinces bordering Hau Giang: Soc Trang, Bac Lieu, Kien Giang, Can Tho, Vinh Long

hau_giang_data = df[df['Province'].isin(['Soc Trang', 'Bac Lieu', 'Kien Giang', 'Can Tho', 'Vinh Long'])].copy()
hau_giang_data.drop(columns=['Province'], inplace=True)
hau_giang_data = hau_giang_data.groupby(by='Date', as_index=False).mean().round(decimals=1)
hau_giang_data['Province'] = 'Hau Giang'
hau_giang_data.head()
```

```python
df = pd.concat(objs=[df, dien_bien_data, dak_nong_data, hau_giang_data], ignore_index=True)
```

```python
df['Province'].nunique()
```

```python
df['Sunrise'] = df['Sunrise'].dt.time
df['Sunset'] = df['Sunset'].dt.time
df.head()
```

```python
df['Country'] = 'Vietnam'
```

```python
df.to_csv(path_or_buf='final_data.csv', index=False)
```
